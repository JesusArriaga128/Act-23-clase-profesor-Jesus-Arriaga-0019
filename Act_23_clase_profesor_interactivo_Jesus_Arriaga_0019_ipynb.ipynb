```python
class Profesor:
    def __init__(self, nombre, materia, antiguedad):
        self.nombre = nombre
        self.materia = materia
        self.antiguedad = antiguedad

    def explicar_clase(self):
        return f"📢 {self.nombre} está explicando {self.materia}."

    def calificar_examen(self):
        return f"✍️ El profesor está calificando exámenes de {self.materia}."

# --- Entrada de datos para Profesor 1 ---
print("--- Datos del Primer Profesor ---")
nom1 = input("Nombre: ")
mat1 = input("Materia: ")
ant1 = input("Antigüedad: ")
profesor1 = Profesor(nom1, mat1, ant1) # Creación del objeto 1

print("\n--- Datos del Segundo Profesor ---")
nom2 = input("Nombre: ")
mat2 = input("Materia: ")
ant2 = input("Antigüedad: ")
profesor2 = Profesor(nom2, mat2, ant2) # Creación del objeto 2

# --- Mostrar resultados ---

print("\n" + "═" * 30)
print(f"🏫 FICHA DOCENTE: {profesor1.nombre}")
print(f"Materia: {profesor1.materia} | Antigüedad: {profesor1.antiguedad}")
print(profesor1.explicar_clase())

print("\n" + "═" * 30)
print(f"🏫 FICHA DOCENTE: {profesor2.nombre}")
print(f"Materia: {profesor2.materia} | Antigüedad: {profesor2.antiguedad}")
print(profesor2.calificar_examen())
```
