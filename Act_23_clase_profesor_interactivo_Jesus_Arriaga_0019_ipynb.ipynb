```python
class Profesor:
    def __init__(self, nombre, materia, antiguedad):
        self.nombre = nombre
        self.materia = materia
        self.antiguedad = antiguedad

    def explicar_clase(self):
        return f"📢 {self.nombre} está explicando un tema de {self.materia}."

    def calificar_examen(self):
        return f"✍️ El profesor está calificando exámenes de {self.materia}."

    def pasar_lista(self):
        return f"✅ {self.nombre} está pasando lista. ¡Digan presente!"

# --- Entrada de datos por el usuario ---

print("📝 Registro de Nuevo Profesor")
nombre_input = input("Ingrese el nombre del profesor: ")
materia_input = input("Ingrese la materia que imparte: ")
antiguedad_input = input("¿Cuántos años de antigüedad tiene?: ")

# --- Creación del Objeto con los datos ingresados ---

nuevo_profesor = Profesor(nombre_input, materia_input, antiguedad_input)

# --- Ejecución de resultados ---

print(f"\n--- 🏫 FICHA DEL DOCENTE GENERADA ---")
print(f"Nombre: {nuevo_profesor.nombre}")
print(f"Materia: {nuevo_profesor.materia}")
print(f"Experiencia: {nuevo_profesor.antiguedad} años")

print("\n--- Acciones en ejecución ---")
print(nuevo_profesor.explicar_clase())
print(nuevo_profesor.pasar_lista())
print(nuevo_profesor.calificar_examen())
```
