```python
class Profesor:
    def __init__(self, nombre, materia, antiguedad):
        self.nombre = nombre
        self.materia = materia
        self.antiguedad = antiguedad

    def explicar_clase(self):
        return f"📢 {self.nombre} está explicando {self.materia}."

# Lista para almacenar los objetos de tipo Profesor
lista_profesores = []

# Usamos un ciclo para pedir los datos dos veces
for i in range(1, 3):
    print(f"\n--- 📝 Ingresando datos del Profesor {i} ---")
    nombre = input(f"Nombre del profesor {i}: ")
    materia = input(f"Materia: ")
    antiguedad = input(f"Antigüedad: ")

    # Creamos el objeto y lo guardamos en la lista
    nuevo_profe = Profesor(nombre, materia, antiguedad)
    lista_profesores.append(nuevo_profe)

# --- Mostrar los resultados ---
print("\n" + "="*30)
print("🏫 LISTADO DE DOCENTES REGISTRADOS")
print("="*30)

for profe in lista_profesores:
    print(f"\n👨‍🏫 Maestro: {profe.nombre}")
    print(f"📚 Imparte: {profe.materia}")
    print(f"⏳ Experiencia: {profe.antiguedad}")
    print(profe.explicar_clase())
```
